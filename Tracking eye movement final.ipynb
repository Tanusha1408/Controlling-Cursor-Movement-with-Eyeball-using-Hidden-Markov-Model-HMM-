```python
import cv2
import mediapipe as mp
import pyautogui
import math
import time

# Initialize Mediapipe FaceMesh
mp_face_mesh = mp.solutions.face_mesh.FaceMesh()

# Open the webcam
cap = cv2.VideoCapture(0)

# Outside the loop
smooth_factor = 0.05
previous_cursor = (0, 0)
blink_counter = 0
blink_threshold = 10  # Adjust as needed
blink_detected = False
right_click_detected = False  # Flag to track if right-click event is detected
event_detected = False  # Flag to track if any event is detected
event_start_time = 0  # Variable to store the time when the event is detected

# Blink smoothing parameters
window_size = 5
blink_history = [0] * window_size

while True:
    # Read the next frame from the video capture
    ret, frame = cap.read()

    # Convert the frame to RGB
    rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)

    # Detect the eyes in the frame
    faces = mp_face_mesh.process(rgb_frame)

    if faces.multi_face_landmarks:
        face = faces.multi_face_landmarks[0]  # Assuming one face for simplicity
        left_eye = face.landmark[33]  # Use the appropriate index for the left eye

        # Calculate the center of the left eye
        left_eye_center = (int(left_eye.x * frame.shape[1]), int(left_eye.y * frame.shape[0]))

        # Calculate the changes in both X and Y directions
        cursor_movement_x = int(smooth_factor * (left_eye_center[0] - previous_cursor[0]))
        cursor_movement_y = int(smooth_factor * (left_eye_center[1] - previous_cursor[1]))

        # Print tracked cursor movements in both X and Y directions
        print("Cursor Movement (X):", cursor_movement_x)
        print("Cursor Movement (Y):", cursor_movement_y)

        # Update the previous cursor position
        previous_cursor = left_eye_center

        # Blink detection
        eye_distance = math.sqrt(cursor_movement_x ** 2 + cursor_movement_y ** 2)
        print("Eye Distance:", eye_distance)

        # Apply moving average filter to blink detection
        blink_history.pop(0)
        blink_history.append(eye_distance)
        smoothed_blink = sum(blink_history) / window_size

        if smoothed_blink < blink_threshold:
            blink_counter += 1
        else:
            blink_counter = 0

        # Check if a blink event occurred
        if blink_counter >= 1:  # Adjust the threshold as needed
            print("Blink detected! Performing a select event.")
            pyautogui.click()
            blink_detected = True
            event_detected = True  # Set the flag to True when an event is detected
            event_start_time = time.time()  # Record the time when the event is detected

        # Print blink detection count
        print("Blink Counter:", blink_counter)

        # Right-click detection (example: closing both eyes)
        if len(faces.multi_face_landmarks) > 1:  # Assuming both eyes are detected
            right_click_detected = True

        # Inside the loop
        cursor_x = int(previous_cursor[0] + cursor_movement_x)
        cursor_y = int(previous_cursor[1] + cursor_movement_y)

        # Move the mouse cursor to the calculated position
        pyautogui.moveTo(cursor_x, cursor_y)
        
        if event_detected:  # Check if any event is detected
            # Perform actions after event detection here
            # Reset event detection flags after 5 seconds
            if time.time() - event_start_time >= 5:  # Adjust the time as needed
                event_detected = False
                blink_detected = False
                right_click_detected = False

        if right_click_detected:
            # Simulate a right-click event
            pyautogui.rightClick()
            print("Right-click detected! Performing a right-click event.")
            event_detected = True
            event_start_time = time.time()  # Record the time when the event is detected

    # Display the frame on the screen
    cv2.imshow("Eye Controlled Mouse", frame)

    # Check for spacebar press to release the mouse and stop the program
    key = cv2.waitKey(1) & 0xFF
    if key == ord(' '):
        # Release the mouse and perform additional actions
        pyautogui.mouseUp()
        # Move the mouse back to its original position
        pyautogui.moveTo(previous_cursor[0], previous_cursor[1])
        # Click the mouse
        pyautogui.click()
        # Release the mouse
        pyautogui.mouseUp()
        break

# Close the video capture
cap.release()

# Close all windows
cv2.destroyAllWindows()
```
